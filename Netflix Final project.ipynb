## ðŸ“Œ1.Univariate Data Visualization

```python
## 1.Which type of content is more prevalent on the platform: Movies or TV Show?


plt.figure(figsize =(5,5))
type_counts = df['type'].value_counts()
plt.pie(type_counts,labels = type_counts.index,autopct='%1.1f%%', startangle=90)
plt.title('Movies vs TV Show Distribution')
plt.show()

type_counts
```

#### Insight:
    â€¢	Movies dominate the Netflix catalog, making up 73.2% of all content, while TV Shows account for 26.8%.
	â€¢	Indicates Netflix prioritizes films for casual viewing, but maintains a significant TV Show library for binge-watchers.
	â€¢	Shows the platformâ€™s content strategy: balancing quick-view movies with longer-form series.
	â€¢	Strategic takeaway: Netflix could consider expanding TV Shows to enhance long-form engagement or continue emphasizing movies to cater to casual viewers.

```python
## 2.Top 10 Directors by Number of Titles


df_net = df[df['director'] != 'Not Available']
top_directors = df_net['director'].value_counts().head(10)
sns.barplot(y=top_directors.index, x=top_directors.values,palette='Paired')
plt.title('Top 10 Directors by Number of Titles')
plt.show()
top_directors
```

#### Insight:
    â€¢	A few prolific directors contribute a significant portion of Netflix content.
	â€¢	RaÃºl Campos and Jan Suter lead with 18 titles, followed by Marcus Raboy (15) and Jay Karas (14).
	â€¢	Directors like Steven Spielberg and Martin Scorsese also appear, showing a mix of international and renowned filmmakers.
	â€¢	Netflix relies on repeat collaborations with these directors to maintain a consistent volume and quality of content.
	â€¢	Strategic takeaway: Focusing on key directors can help Netflix curate popular content, plan marketing strategies, and negotiate futureÂ projects.

```python
## 3.Which countries produce the most content?Â¶

top_country = df['country'].value_counts().head(10)
sns.barplot(y = top_country.index ,x = top_country.values, palette='Set2' )
plt.title('Top 10 Country Created Content')
plt.tight_layout()
plt.show()
top_country
```

	â€¢	Other notable contributors include United Kingdom, Japan, and South Korea.
	â€¢	Reflects Netflixâ€™s global strategy and focus on regional content.
	â€¢	Strategic takeaway: Understanding production hubs helps guide licensing, localization, and market targeting.

```python
## 4.Distribution of movie duration

plt.figure(figsize=(10,6))
plt.title("Distribution of Movie Durations (Minutes)")
sns.histplot(movie_durations, bins=40, kde=False, color="skyblue")
plt.ylabel("Number of movies")
plt.xlabel("Duration (Minutes)")
plt.show()
```

```python
## 5.What are the most common ratings for movies and TV shows?

plt.figure(figsize=(7,5))
rating_order = df['rating'].value_counts().index
sns.countplot(x='rating', data=df , order = rating_order,palette='Set1')
plt.xticks(rotation=45)
plt.title('Distribution of Content Ratings on Netflix')
plt.xlabel('Rating')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

#### Insight:

## ðŸ“Œ2.Bivariate Data Visualization

```python
## 1.What are the most common content ratings for each type of content (movies and TV shows)?

plt.figure(figsize=(8, 5))
sns.countplot(x='rating', hue='type', data=df)
plt.title('Content Type vs. Rating')
plt.xticks(rotation=90)
plt.show()
```

#### Insight:
    â€¢	TV-MA, TV-14, and PG-13 are the most common ratings across Netflix content.
	â€¢	Movies and TV Shows primarily target adult and teen audiences, with less content for children (G, TV-Y).
	â€¢	Shows Netflixâ€™s content suitability strategy for different age groups.
	â€¢	Strategic takeaway: Helps guide audience targeting, parental guidance, and content recommendations.
   

```python
## 2.Movies vs TV Shows Over the Years

type_year = df.groupby(['release_year', 'type']).size().unstack().sort_index()
plt.figure(figsize=(6, 4))
type_year.plot(kind='line', marker='o', linewidth=2, figsize=(12, 6), color=['red', 'green'])
plt.title('Number of Movies vs TV Shows Released Over the Years', fontsize=14)
plt.xlabel('Release Year', fontsize=12)
plt.ylabel('Number of Titles', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.xticks(rotation=45)
plt.legend(title='Content Type')
plt.tight_layout()
plt.show()
```

#### Insight:
    â€¢	Movies consistently outnumber TV Shows across most release years.
	â€¢	Both content types show gradual growth, with occasional spikes in certain years.
	â€¢	Indicates Netflixâ€™s focus on movies for casual viewers, while steadily expanding TV Shows for binge-watchers.
	â€¢	Strategic takeaway: Understanding yearly trends helps Netflix plan acquisitions and content strategy.

```python
## 3.Correlation between two numeric columns

corr_data = df[['release_year','year']]
correlation = corr_data.corr()
print("Correlation between release_year and year:")
print(correlation)
plt.figure(figsize=(5,4))
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation: Release Year vs Year Added')
plt.show()
```

#### Insight:
    â€¢  	Correlation between release year and year added is very weak (0.072).
	â€¢	Indicates Netflix adds both old and new content, not strictly based on release year.
	â€¢	Shows content addition is influenced more by licensing and demand than by release recency.

```python
## 4.Top 10 Countries vs. Type (Movie/TV Show)

top_10_countries = df['country'].value_counts().head(10).index
df_top_countries = df[df['country'].isin(top_10_countries)]
plt.figure(figsize=(10, 5))
sns.countplot(x='country', hue='type', data=df_top_countries,order = top_10_countries)
plt.title('Top 10 Countries vs. Type (Movie/TV Show)')
plt.xticks(rotation=45)
plt.xlabel('Country')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

#### Insight:
    â€¢	United States, India, and United Kingdom lead content production among the top 10 countries.
	â€¢	Both Movies and TV Shows are produced, though some countries focus more on one type.
	â€¢	Reflects Netflixâ€™s global expansion and regional content strategy.
	â€¢	Strategic takeaway: Identifying top producing countries helps guide licensing, localization, and market targeting.

```python
## 5.Are TV shows added in different years compared to movies?


plt.figure(figsize=(14, 6))
sns.countplot(x='year', hue='type', data=df,palette='Set1')
plt.title('Number of Movies and TV Shows per Year')
plt.xlabel('Year')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
## 4.Rating Ã— Genre (Which genres get which ratings?)


# Extract the first listed genre
df['main_genre'] = df['listed_in'].str.split(',').str[0]

plt.figure(figsize=(14,7))
sns.countplot(
    x='rating',
    hue='main_genre',
    data=df,
    order=df['rating'].value_counts().index
)
plt.xticks(rotation=90)
plt.title("Ratings by Main Genre")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.legend(title='Genre', fontsize=9, loc='upper right')
plt.tight_layout()
plt.show()
```

	â€¢	TV-14 targets teens/young adults with Romance, Comedy, Action, and Teen Dramas.
